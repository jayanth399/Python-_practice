# Null value treatment and imputation 
A.Fill the missing records:<br
    
B.Remove the missing values:<br>
1.Remove the rows in which missing position <br>
2.Remove the columns in which missing positions <br>

```python
(df.isna().sum()/df.shape[0])*100
```

```python
df.fillna({'age': df['age'].median()}, inplace=True)
```

```python
(df.isna().sum() / df.shape[0]) * 100
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```

```python
(df.isna().sum() / df.shape[0]) * 100
```

```python
df.drop(columns=['embarked'], inplace=True)
```

```python
(df.isna().sum() / df.shape[0]) * 100
```
